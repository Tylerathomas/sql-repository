```python
from sqlalchemy import create_engine
import pandas as pd
import mysql.connector
%load_ext sql

engine = create_engine('mysql://USERNAME:PASSWORD@HOST/DATABASE')
```

**The purpose of Query 1 is to retrieve the total number of reviews for each of the three apps - Twitter, Anydo and TikTok - from their respective tables in the database. The UNION ALL operator is used to combine the results of the three SELECT statements into a single table with columns for app name and total number of reviews.**

```python
# Query 1 - Check the total number of reviews for each app.
query1 = """
SELECT 'Twitter' AS app_name, COUNT(*) AS total_reviews FROM `Twitter_Reviews`
UNION ALL
SELECT 'Anydo' AS app_name, COUNT(*) AS total_reviews FROM `anydo_playstore_reviews`
UNION ALL
SELECT 'TikTok' AS app_name, COUNT(*) AS total_reviews FROM `tiktok_google_play_reviews`;
"""
df1 = pd.read_sql_query(query1, engine)
display(df1)
```

The insight from this query is that TikTok has the highest number of reviews among the three apps, while Anydo has the lowest. The total number of reviews can give an indication of the popularity of the app and the level of engagement from its users.

**This query is finding the average thumbs up count and average rating for each app by app_name. It uses a UNION ALL to combine the results for each app into a single dataframe.**

```python
# Query 2 - Find the average rating and the average thumbs up count for each app.
query2 = """
SELECT 'Twitter' AS app_name, AVG(`Thumbs_Up_Count`) AS avg_thumbs_up, NULL AS avg_rating FROM `Twitter_Reviews`
UNION ALL
SELECT 'Anydo' AS app_name, AVG(`thumbsUpCount`) AS avg_thumbs_up, AVG(`score`) AS avg_rating FROM `anydo_playstore_reviews`
UNION ALL
SELECT 'TikTok' AS app_name, AVG(`thumbsUpCount`) AS avg_thumbs_up, AVG(`score`) AS avg_rating FROM `tiktok_google_play_reviews`;
"""
df2 = pd.read_sql_query(query2, engine)
display(df2)
```

The output shows the average thumbs up count and average rating for each app in separate columns. The Twitter app does not have a rating, so the column for avg_rating shows NULL. The Anydo and Tiktok apps have both an average thumbs up count and an average rating.

The insight we can gain from this query is that the Anydo and Tiktok apps have similar average thumbs up counts, but the Anydo app has a slightly higher average rating than the Tiktok app. The Twitter app has a much lower average thumbs up count than the other two apps, which suggests that it may not be as popular or well-liked as the other two.

**After running the query, we can see that Twitter has the highest percentage of reviews with thumbs-up count greater than or equal to 10 (30.05%), followed by Anydo (6.87%) and TikTok (0.17%). This insight is useful for understanding the level of engagement and satisfaction among users of each app, and can help the app developers identify areas for improvement.**

```python
#Query 3 - Find the percentage of reviews with thumbs-up count greater than or equal to 10 
query3 = """
SELECT app_name, 
       SUM(CASE WHEN thumbs_up >= 10 THEN 1 ELSE 0 END) / COUNT(*) * 100 AS percentage_of_reviews_with_10_or_more_thumbs_up
FROM (
    SELECT 'Twitter' AS app_name, `Thumbs_Up_Count` AS thumbs_up FROM `Twitter_Reviews`
    UNION ALL
    SELECT 'Anydo' AS app_name, `thumbsUpCount` AS thumbs_up FROM `anydo_playstore_reviews`
    UNION ALL
    SELECT 'TikTok' AS app_name, `thumbsUpCount` AS thumbs_up FROM `tiktok_google_play_reviews`
) AS subquery
GROUP BY app_name;
"""
df3 = pd.read_sql_query(query3, engine)
display(df3)
```

After running the query, we can see that Twitter has the highest percentage of reviews with thumbs-up count greater than or equal to 10 (30.05%), followed by Anydo (6.87%) and TikTok (0.17%). This insight is useful for understanding the level of engagement and satisfaction among users of each app, and can help the app developers identify areas for improvement.

**Query 4 creates a view called all_app_reviews_word_split that splits the review content into individual words. Then, it finds the most common words used in the reviews for each app, up to the top 5 per platform. The query first creates a CTE word_counts that calculates the count of each word used in the reviews for each app, and then assigns a row number to each group of words based on the count. The main query selects the words with the top 5 counts for each app, and orders them by the app name and row number.**

```python
# Query 4 - Create a view that splits the review content into individual words
drop_view_sql_word_split = "DROP VIEW IF EXISTS all_app_reviews_word_split;"
with engine.connect() as con:
    con.execute(drop_view_sql_word_split)

create_view_sql_word_split = """
CREATE VIEW all_app_reviews_word_split AS
SELECT app_name,
       User_Name,
       Comment,
       Thumbs_Up_Count,
       Version,
       created_at,
       SUBSTRING_INDEX(SUBSTRING_INDEX(Comment, ' ', n), ' ', -1) AS word
FROM all_app_reviews
JOIN (SELECT 1 n UNION ALL SELECT 2 UNION ALL SELECT 3 UNION ALL SELECT 4 UNION ALL SELECT 5 UNION ALL SELECT 6 UNION ALL SELECT 7 UNION ALL SELECT 8 UNION ALL SELECT 9 UNION ALL SELECT 10) numbers
ON CHAR_LENGTH(Comment) - CHAR_LENGTH(REPLACE(Comment, ' ', '')) >= n - 1;
"""

with engine.connect() as con:
    con.execute(create_view_sql_word_split)

# Query 4 - Find the most common words used in the reviews for each app (top 5 per platform)
query4 = """
WITH word_counts AS (
  SELECT app_name, word, COUNT(*) AS word_count,
         ROW_NUMBER() OVER (PARTITION BY app_name ORDER BY COUNT(*) DESC) AS row_num
  FROM all_app_reviews_word_split
  WHERE word NOT IN ('', 'the', 'and', 'to', 'a', 'of', 'in', 'for', 'is', 'on', 'with', 'I', 'it', 'that', 'this', 'as', 'are', 'have', 'was', 'but', 'my', 'be', 'at', 'not', 'or', 'by', 'so', 'if', 'from', 'an', 'there', 'has', 'about', 'you', 'The', 'can', 'will', 'its', 'their', 'they', 'them', 'we', 'our', 'us', 'your', 'when', 'would', 'should', 'how', 'who', 'what', 'where', 'which', 'It''s','me','app','very','really','app.','twitter','tiktok')
  GROUP BY app_name, word
)
SELECT app_name, word, word_count
FROM word_counts
WHERE row_num <= 5
ORDER BY app_name, row_num;
"""

df4 = pd.read_sql_query(query4, engine)
display(df4)

```

The fourth query creates a view that splits the review content into individual words, and then finds the most common words used in the reviews for each app (top 5 per platform). We can see that the word "good" is the most common in Anydo and TikTok reviews, while "good" and "update" are the most common in Twitter reviews. This information can be useful for each company to gain insights into their customer's preferences and needs.

**This query calculates the number of reviews for each app that fall into three categories based on the number of thumbs up counts: low (0-9), medium (10-19), and high (>=20). It uses a common table expression (CTE) to first categorize each review based on the thumbs up count, and then the main query aggregates the number of reviews for each app in each category.**

```python
# Query 5 - Find the number of reviews per app that have a high (>= 20), medium (10-19), or low (0-9) thumbs-up count
query5 = """
WITH thumbs_up_categories AS (
    SELECT app_name,
           Thumbs_Up_Count AS thumbs_up,
           CASE
               WHEN Thumbs_Up_Count >= 20 THEN 'High'
               WHEN Thumbs_Up_Count BETWEEN 10 AND 19 THEN 'Medium'
               ELSE 'Low'
           END AS thumbs_up_category
    FROM all_app_reviews
)
SELECT app_name,
       thumbs_up_category,
       COUNT(*) AS num_reviews
FROM thumbs_up_categories
GROUP BY app_name, thumbs_up_category
ORDER BY app_name, thumbs_up_category DESC;
"""

df5 = pd.read_sql_query(query5, engine)
display(df5)
```

We can see that Anydo has the highest number of low thumbs-up count reviews with 8028, followed by Twitter with 838 and TikTok with 1813. Twitter has the highest number of high thumbs-up count reviews with 270, followed by Anydo with 336 and TikTok with 3.

**The purpose of Query 6 is to identify the top 5 most frequently mentioned keywords in positive and negative comments for each app. The query creates a view named "comment_words" which splits the comment column of each app into individual words. The query then creates another view named "word_counts" which counts the number of positive and negative comments for each word in the comment_words view. Finally, the query selects the top 5 most frequently mentioned keywords in positive and negative comments for each app by using the RANK() function to assign a rank to each keyword based on the positive and negative counts.**

```python
# Query 6 - Identify the top 5 most frequently mentioned keywords in positive and negative comments for each app
query6 = """
WITH comment_words AS (
  SELECT app_name,
         Thumbs_Up_Count,
         SUBSTRING_INDEX(SUBSTRING_INDEX(Comment, ' ', n.n), ' ', -1) AS word
  FROM all_app_reviews
  JOIN (
    SELECT 1 AS n
    UNION ALL SELECT 2 UNION ALL SELECT 3 UNION ALL SELECT 4 UNION ALL SELECT 5
    UNION ALL SELECT 6 UNION ALL SELECT 7 UNION ALL SELECT 8 UNION ALL SELECT 9
    UNION ALL SELECT 10
  ) n
  ON CHAR_LENGTH(Comment) - CHAR_LENGTH(REPLACE(Comment, ' ', '')) >= n.n - 1
),
word_counts AS (
  SELECT app_name,
         word,
         SUM(CASE WHEN Thumbs_Up_Count >= 3 THEN 1 ELSE 0 END) AS positive_count,
         SUM(CASE WHEN Thumbs_Up_Count < 3 THEN 1 ELSE 0 END) AS negative_count
  FROM comment_words
  WHERE word NOT IN ('', 'the', 'and', 'to', 'a', 'of', 'in', 'for', 'is', 'on', 'with', 'I', 'it', 'that', 'this', 'as', 'are', 'have', 'was', 'but', 'my', 'be', 'at', 'not', 'or', 'by', 'so', 'if', 'from', 'an', 'there', 'has', 'about', 'you', 'The', 'can', 'will', 'its', 'their', 'they', 'them', 'we', 'our', 'us', 'your', 'when', 'would', 'should', 'how', 'who', 'what', 'where', 'which', 'It''s', 'me', 'app', 'very', 'really', 'app.', 'twitter', 'tiktok')
  GROUP BY app_name, word
)
SELECT app_name,
       word,
       positive_count,
       negative_count
FROM (
  SELECT *,
         ROW_NUMBER() OVER (PARTITION BY app_name ORDER BY positive_count DESC) AS positive_rank,
         ROW_NUMBER() OVER (PARTITION BY app_name ORDER BY negative_count DESC) AS negative_rank
  FROM word_counts
) ranked_words
WHERE positive_rank <= 5 OR negative_rank <= 5
ORDER BY app_name, positive_count DESC, negative_count DESC;
"""
df6 = pd.read_sql_query(query6, engine)
display(df6)
```

We can see that "love", "like", and "good" are commonly mentioned in positive reviews for Anydo. In negative reviews, "app" is commonly mentioned. For TikTok, we can see that "good" is mentioned in one positive review, and "don't" and "Post" are mentioned in one negative review each. For Twitter, "update" and "New" are commonly mentioned in positive reviews, while "app" is mentioned frequently in negative reviews.

**This query retrieves the distribution of review scores for each app in the dataset. The result includes the app name, review score, and the number of reviews for each score.**

```python
# Query 7 - Distribution of review scores for each app
query7 = """
SELECT app_name,
       score AS review_score,
       COUNT(*) AS num_reviews
FROM all_app_reviews
GROUP BY app_name, review_score
ORDER BY app_name, review_score;
"""
df7 = pd.read_sql_query(query7, engine)
display(df7)
```

The insight from this query is the distribution of review scores for each app in the dataset. We can analyze the results to determine the overall sentiment of the reviews for each app.

**The query defined above creates a view named 'all_app_reviews' that consolidates the review data from multiple tables into a single view. It combines the relevant columns from the 'Twitter_Reviews', 'anydo_playstore_reviews', and 'tiktok_google_play_reviews' tables, and adds an additional column 'app_name' to identify the source app for each review.**

Query 8 utilizes the 'all_app_reviews' view to find the app versions with the most positive and negative feedback. It first calculates the sentiment of each review based on the score, categorizing them as 'positive', 'negative', or 'neutral'. Then, it counts the number of reviews for each app version and sentiment. Finally, it selects the app versions with the highest count for positive and negative sentiments for each app.

```python
# Define the query to create the view 'all_app_reviews'
create_view_query = """
CREATE OR REPLACE VIEW all_app_reviews AS
SELECT
  Review_Id AS review_id,
  User_Name AS user_name,
  User_Image AS user_image,
  Comment AS content,
  NULL AS score,  -- Placeholder value for score column in Twitter_Reviews
  Thumbs_Up_Count AS thumbs_up_count,
  Version AS review_created_version,
  Review_Created_At AS at,
  Sort_Order AS sort_order,
  'Twitter' AS app_name
FROM Twitter_Reviews
UNION ALL
SELECT
  reviewId AS review_id,
  userName AS user_name,
  userImage AS user_image,
  content,
  score,
  thumbsUpCount AS thumbs_up_count,
  reviewCreatedVersion AS review_created_version,  -- Correct column name for app version
  at,
  sortOrder AS sort_order,
  'Anydo' AS app_name
FROM anydo_playstore_reviews
UNION ALL
SELECT
  reviewId AS review_id,
  userName AS user_name,
  userImage AS user_image,
  content,
  score,
  thumbsUpCount AS thumbs_up_count,
  reviewCreatedVersion AS review_created_version,  -- Correct column name for app version
  at,
  NULL AS sort_order,
  'TikTok' AS app_name
FROM tiktok_google_play_reviews;
"""

# Execute the query to create the view
engine.execute(create_view_query)

# Define Query 8 to find app versions with the most positive and negative feedback
query8 = """
WITH version_sentiment AS (
  SELECT app_name,
         review_created_version AS app_version,
         CASE 
           WHEN score >= 4 THEN 'positive'
           WHEN score <= 2 THEN 'negative'
           ELSE 'neutral'
         END AS sentiment,
         COUNT(*) AS sentiment_count
  FROM all_app_reviews
  WHERE review_created_version IS NOT NULL  -- Exclude rows with no app_version
  GROUP BY app_name, app_version, sentiment
)
SELECT app_name,
       app_version,
       sentiment,
       sentiment_count
FROM (
  SELECT *,
         RANK() OVER (PARTITION BY app_name, sentiment ORDER BY sentiment_count DESC) AS sentiment_rank
  FROM version_sentiment
) ranked
WHERE sentiment_rank = 1
ORDER BY app_name, sentiment;
"""

# Execute Query 8 and display the result
df8 = pd.read_sql_query(query8, engine)
display(df8)
```

The result of Query 8 displays the app name, app version, sentiment ('positive' or 'negative'), and the count of reviews with that sentiment for the respective app versions. This information helps identify the specific versions of the apps that have received the most positive and negative feedback.
